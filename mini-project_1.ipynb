Исправьте ситуацию: приведите названия столбцов к нижнему регистру и замените пробелы на знак нижнего подчёркивания. Например: Is Canceled -> is_canceled

Могут пригодиться:

* str.lower

* columns

Сделал максимально прямолинейно, но работает

```python
bookings =bookings.rename(columns={"Hotel": "hotel", "Is Canceled": "is_canceled", \
                         "Lead Time": "lead_time", "arrival full date": "arrival_full_date",\
                         "Is Canceled": "is_canceled", "Babies": "babies" \
                        , "Arrival Date Year": "arrival_date_year", "Arrival Date Month": "arrival_date_month"\
                        , "Arrival Date Week Number": "arrival_date_week_number", "Arrival Date Day of Month": "arrival_date_day_of_month"\
                        , "Stays in Weekend nights": "stays_in_weekend_nights", "Stays in week nights": "stays_in_week_nights"\
                        , "stays total nights": "stays_total_nights" , "Adults": "adults" \
                        , "Children": "children" , "Meal": "meal"\
                        , "Country": "country", "Reserved Room Type": "reserved_room_type"\
                        , "Assigned room type": "assigned_room_type", "customer type": "customer_type" \
                        , "Reservation Status": "reservation_status" , "Reservation status_date": "reservation_status_date"})
```

Попробую более разумный вариант

```python
rename_col={}
for col in bookings.columns:
    rename_col[col]=col.replace(" ", "_").lower()
```

```python
rename_col
```

```python
bookings = bookings.rename(columns=rename_col)
```
